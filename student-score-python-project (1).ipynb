```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#READING THE FILE

df = pd.read_csv("/kaggle/input/students-exam-scores/Expanded_data_with_more_features.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#Pre-processing 
#Drop unnamed column
```

```python
df = df.drop("Unnamed: 0", axis = 1)
df.head()
```

```python
#change weekly study hours column
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

```python
#Gender Distribution
```

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data = df,x = "Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#From the above chart we have analyzed that no. of females are more than males.
```

```python
#Parent Education impact on different Score
```

```python
gp = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore": 'mean',"WritingScore":'mean'})
print(gp)  
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gp,annot=True)
plt.show()
```

```python
#from the above chart we have concluded that the education of the parents have a good impact on their scores.
```

```python
#groupby based on parent marital status 
```

```python
gp1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore": 'mean',"WritingScore":'mean'})
print(gp1)  
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gp1,annot=True)
plt.show()
```

```python
#From the above chart we have concluded that there is no/negligible impact on the student score due to their parents marital status.
```

```python
groupA = df.loc[(df["EthnicGroup"]=="group A")].count()
print(groupA)
```

```python
groupB = df.loc[(df["EthnicGroup"]=="group B")].count()
groupC = df.loc[(df["EthnicGroup"]=="group C")].count()
groupD = df.loc[(df["EthnicGroup"]=="group D")].count()
groupE = df.loc[(df["EthnicGroup"]=="group E")].count()
```

```python
l = ["groupA","groupB","groupC","groupD","groupE"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist, labels = l, autopct='%1.2f%%')
plt.title("Distribution of Ethnic Groups")
plt.show()
```
